```python

import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_burglaries = pd.read_csv(r'C:\Users\20202719\Documents\TUE\Y2 DS\Q4\Data challenge 2\DataChallenge2\DC2Data\full_data.csv')

df_burglaries.head()
```

```python
df_burglaries['Month of crime'] = df_burglaries['Month_x'].str[5:7]
df_burglaries['Year of crime'] = df_burglaries['Month_x'].str[:4]

df_burglaries['Number of burglaries'] = 1



f = {
    'Mid-year Population Estimates;All Ages;2013' : 'mean',
    'Mid-year Population Estimates;Aged 0-15;2013' : 'mean',
    'Mid-year Population Estimates;Aged 16-29;2013' : 'mean',
    'Mid-year Population Estimates;Aged 30-44;2013' :'mean',
    'Mid-year Population Estimates;Aged 45-64;2013' : 'mean',
    'Mid-year Population Estimates;Aged 65+;2013' : 'mean',
    'Population Density;Persons per hectare;2013' : 'mean',
    'Ethnic Group;White;2011' : 'mean',
    'Ethnic Group;Mixed/multiple ethnic groups;2011' : 'mean',
    'Ethnic Group;Asian/Asian British;2011' : 'mean',
    'Ethnic Group;Black/African/Caribbean/Black British;2011' : 'mean',
    'Ethnic Group;Other ethnic group;2011' : 'mean',
    'Number of burglaries' : 'sum',
    'Household Income, 2011/12;Median Annual Household Income estimate (£)' : 'mean'

}


df_burglaries_grouped = df_burglaries.groupby(['LSOA code_x', 'Year of crime']).agg(f)
df_burglaries_grouped = df_burglaries_grouped.reset_index()


df_burglaries_grouped['LSOA code reg'] = df_burglaries_grouped['LSOA code_x'].str[1:]

key_list = []
for key,value in f.items():
    key_list.append(key)


covid_mask = (df_burglaries_grouped['Year of crime'] != '2020') & (df_burglaries_grouped['Year of crime'] != '2021') & (df_burglaries_grouped['Year of crime'] != '2022')
df_burglaries_woCovid_grouped = df_burglaries_grouped[covid_mask]

df_burglaries_woCovid_grouped.head()
```

```python
#Training the model
regr = LinearRegression()

X = df_burglaries_woCovid_grouped[['LSOA code reg',
 'Mid-year Population Estimates;All Ages;2013',
 'Mid-year Population Estimates;Aged 0-15;2013',
 'Mid-year Population Estimates;Aged 16-29;2013',
 'Mid-year Population Estimates;Aged 30-44;2013',
 'Mid-year Population Estimates;Aged 45-64;2013',
 'Population Density;Persons per hectare;2013',
 'Ethnic Group;White;2011',
 'Ethnic Group;Mixed/multiple ethnic groups;2011',
 'Ethnic Group;Asian/Asian British;2011',
 'Ethnic Group;Black/African/Caribbean/Black British;2011',
 'Household Income, 2011/12;Median Annual Household Income estimate (£)',
 'Year of crime'
 ]]
y = df_burglaries_woCovid_grouped['Number of burglaries']




X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)

#regr.fit(X_train,y_train)

#Y_hat = regr.predict(X_test)

#print(regr.intercept_, regr.coef_, regr.score(X_test, y_test))

```

```python
#Random forest model
model = RandomForestClassifier()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```
